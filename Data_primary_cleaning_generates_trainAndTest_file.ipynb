```python
data['操作时间'][data['操作时间'].isnull()]='10-10月-16'
data['操作时间'][data['操作时间']=='200000']='10-10月-16'#处理脏数据
data['操作时间new'] = data['操作时间'].apply(lambda x:'20'+str(x).split('-')[2]+'-'+str(x).split('-')[1].split('月')[0]+'-'+str(x).split('-')[0])
data['操作时间_月'] = pd.DatetimeIndex(data['操作时间new']).month
data['操作时间_dayofweek'] = pd.DatetimeIndex(data['操作时间new']).dayofweek

data['住院开始时间'][data['住院开始时间'].isnull()]='10-10月-16'
data['住院开始时间'][data['住院开始时间']=='200000']='10-10月-16'#处理脏数据
data['住院开始时间new'] = data['操作时间'].apply(lambda x:'20'+str(x).split('-')[2]+'-'+str(x).split('-')[1].split('月')[0]+'-'+str(x).split('-')[0])
data['住院开始时间_月'] = pd.DatetimeIndex(data['住院开始时间new']).month
data['住院开始时间_dayofweek'] = pd.DatetimeIndex(data['住院开始时间new']).dayofweek

data['交易时间_月'] = pd.DatetimeIndex(data['交易时间']).month
data['交易时间_dayofweek'] = pd.DatetimeIndex(data['交易时间']).dayofweek

data
```

```python
data[['操作时间_月']] = data[['操作时间_月']] .astype('float64')
data[['操作时间_dayofweek']] = data[['操作时间_dayofweek']] .astype('float64')
data[['住院开始时间_月']] = data[['住院开始时间_月']] .astype('float64')
data[['住院开始时间_dayofweek']] = data[['住院开始时间_dayofweek']] .astype('float64')
data[['交易时间_月']] = data[['交易时间_月']] .astype('float64')
data[['交易时间_dayofweek']] = data[['交易时间_dayofweek']] .astype('float64')
```

```python
caozuoMon = pd.get_dummies(data['操作时间_月'],prefix='操作时间_月')
caozuoDay = pd.get_dummies(data['操作时间_dayofweek'],prefix='操作时间_dayofweek')
zhuyuanMon = pd.get_dummies(data['住院开始时间_月'],prefix='住院开始时间_月')
zhuyuanDay = pd.get_dummies(data['住院开始时间_dayofweek'],prefix='住院开始时间_dayofweek')
jiaoyiMon = pd.get_dummies(data['交易时间_月'],prefix='交易时间_月')
jiaoyiDay = pd.get_dummies(data['交易时间_dayofweek'],prefix='交易时间_dayofweek')
```

```python
result = pd.concat([data,caozuoMon],axis=1)
result = pd.concat([result,caozuoDay],axis=1)
result = pd.concat([result,zhuyuanMon],axis=1)
result = pd.concat([result,zhuyuanDay],axis=1)
result = pd.concat([result,jiaoyiMon],axis=1)
result = pd.concat([result,jiaoyiDay],axis=1)
```
