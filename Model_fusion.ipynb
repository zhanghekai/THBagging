```python
#正反样例比例为19比1
#遂欲将0例10份，与1例组成2:1，共10个样本集，分别训练一个分类器
data_tree_train_0 = data_tree_train[data_tree_train.label==0]#0样例 19000个
data_tree_train_1 = data_tree_train[data_tree_train.label==1]#1样例 1000个
data_tree_train_0_lr = data_tree_train_0[:3800]#0样例 3800个 用于lr训练
data_tree_train_0_xgb = data_tree_train_0[3800:]#0样例 15200个 用于xgb训练
data_tree_train_1_lr = data_tree_train_1[:200]#1样例 200个 用于lr训练
data_tree_train_1_xgb = data_tree_train_1[200:]#0样例 800个 用于xgb训练
```

```python
train_0_1 = data_tree_train_0_xgb[:1500]
train_0_2 = data_tree_train_0_xgb[1500:3000]
train_0_3 = data_tree_train_0_xgb[3000:4500]
train_0_4 = data_tree_train_0_xgb[4500:6000]
train_0_5 = data_tree_train_0_xgb[6000:7500]
train_0_6 = data_tree_train_0_xgb[7500:9000]
train_0_7 = data_tree_train_0_xgb[9000:10500]
train_0_8 = data_tree_train_0_xgb[10500:12000]
train_0_9 = data_tree_train_0_xgb[12000:13500]
train_0_10 = data_tree_train_0_xgb[13500:]
```

```python
train_1 = pd.concat([train_0_1,data_tree_train_1_xgb],ignore_index=True)
train_2 = pd.concat([train_0_2,data_tree_train_1_xgb],ignore_index=True)
train_3 = pd.concat([train_0_3,data_tree_train_1_xgb],ignore_index=True)
train_4 = pd.concat([train_0_4,data_tree_train_1_xgb],ignore_index=True)
train_5 = pd.concat([train_0_5,data_tree_train_1_xgb],ignore_index=True)
train_6 = pd.concat([train_0_6,data_tree_train_1_xgb],ignore_index=True)
train_7 = pd.concat([train_0_7,data_tree_train_1_xgb],ignore_index=True)
train_8 = pd.concat([train_0_8,data_tree_train_1_xgb],ignore_index=True)
train_9 = pd.concat([train_0_9,data_tree_train_1_xgb],ignore_index=True)
train_10 = pd.concat([train_0_10,data_tree_train_1_xgb],ignore_index=True)
```

```python
import numpy as np
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
from sklearn import cross_validation, metrics
from sklearn.grid_search import GridSearchCV
import time
```

```python
target='label'
IDcol = '个人编码'
predictors = [x for x in data_tree_train.columns if x not in [target, IDcol]]
```

```python
start = time.time()
param_test1 = {
    'max_depth':np.array([5,10,20]),
    'min_child_weight':np.array([1,3,5,7]),
#     'gamma':[i/10.0 for i in range(0,5)],
#     'n_estimators':np.array([700])
}
gsearch1 = GridSearchCV(estimator = XGBClassifier( learning_rate =0.1, n_estimators=100, max_depth=7,
                                        min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8,
                                        objective= 'binary:logistic', nthread=4, scale_pos_weight=1, seed=27), 
                       param_grid = param_test1, scoring='f1',n_jobs=4,iid=False, cv=5)
gsearch1.fit(train_3[predictors],train_3[target])
end = time.time()
print('运行时间：')
(end-start)/60
```

```python
gsearch1.grid_scores_, gsearch1.best_params_, gsearch1.best_score_
```

```python
#上面不同的测试集，进行了参数选择，不同的参数差别不大，但还是会有
#为了改变超参数而增加blending后模型的随机性和多样性，所以无法使用for循环，只能复制粘贴，手动修改参数
xgb1 = XGBClassifier(learning_rate =0.1, n_estimators=1000,max_depth=7,min_child_weight=1,gamma=0,subsample=0.8,colsample_bytree=0.8,objective= 'binary:logistic',nthread=4,scale_pos_weight=1,seed=27)
xgb1.fit(train_1[predictors], train_1['label'])
xgb2 = XGBClassifier(learning_rate =0.1, n_estimators=100,max_depth=6,min_child_weight=2,gamma=0,subsample=0.8,colsample_bytree=0.8,objective= 'binary:logistic',nthread=4,scale_pos_weight=1,seed=27)
xgb2.fit(train_2[predictors], train_2['label'])
xgb3 = XGBClassifier(learning_rate =0.1, n_estimators=200,max_depth=5,min_child_weight=3,gamma=0,subsample=0.8,colsample_bytree=0.8,objective= 'binary:logistic',nthread=4,scale_pos_weight=1,seed=27)
xgb3.fit(train_3[predictors], train_3['label'])
xgb4 = XGBClassifier(learning_rate =0.1, n_estimators=300,max_depth=4,min_child_weight=4,gamma=0,subsample=0.8,colsample_bytree=0.8,objective= 'binary:logistic',nthread=4,scale_pos_weight=1,seed=27)
xgb4.fit(train_4[predictors], train_4['label'])
xgb5 = XGBClassifier(learning_rate =0.1, n_estimators=400,max_depth=3,min_child_weight=5,gamma=0,subsample=0.8,colsample_bytree=0.8,objective= 'binary:logistic',nthread=4,scale_pos_weight=1,seed=27)
xgb5.fit(train_5[predictors], train_5['label'])
xgb6 = XGBClassifier(learning_rate =0.1, n_estimators=500,max_depth=8,min_child_weight=6,gamma=0,subsample=0.8,colsample_bytree=0.8,objective= 'binary:logistic',nthread=4,scale_pos_weight=1,seed=27)
xgb6.fit(train_6[predictors], train_6['label'])
xgb7 = XGBClassifier(learning_rate =0.1, n_estimators=600,max_depth=9,min_child_weight=7,gamma=0,subsample=0.8,colsample_bytree=0.8,objective= 'binary:logistic',nthread=4,scale_pos_weight=1,seed=27)
xgb7.fit(train_7[predictors], train_7['label'])
xgb8 = XGBClassifier(learning_rate =0.1, n_estimators=700,max_depth=10,min_child_weight=8,gamma=0,subsample=0.8,colsample_bytree=0.8,objective= 'binary:logistic',nthread=4,scale_pos_weight=1,seed=27)
xgb8.fit(train_8[predictors], train_8['label'])
xgb9 = XGBClassifier(learning_rate =0.1, n_estimators=800,max_depth=11,min_child_weight=9,gamma=0,subsample=0.8,colsample_bytree=0.8,objective= 'binary:logistic',nthread=4,scale_pos_weight=1,seed=27)
xgb9.fit(train_9[predictors], train_9['label'])
xgb10 = XGBClassifier(learning_rate =0.1, n_estimators=900,max_depth=12,min_child_weight=10,gamma=0,subsample=0.8,colsample_bytree=0.8,objective= 'binary:logistic',nthread=4,scale_pos_weight=1,seed=27)
xgb10.fit(train_10[predictors], train_10['label'])
```
