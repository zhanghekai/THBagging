```python
import pandas as pd
```

```python
data_tree_train = pd.read_csv('data_tree_train.csv',encoding='gbk')

data_tree_test = pd.read_csv('data_tree_test.csv',encoding='gbk')
data_tree_test['label'] = -1
data = pd.concat([data_tree_train,data_tree_test],ignore_index=True)
data
```

```python
data.isnull().sum().sort_values(ascending=False)
```

```python
data = data.fillna(0)
data.isnull().sum().sort_values(ascending=False)
```

```python
data[['医院编码_min','医院编码_max']] = data[['医院编码_min','医院编码_max']].astype('object')
data = pd.get_dummies(data)
data
```

```python
target='label'
IDcol = '个人编码'
predictors1 = [x for x in data.columns if x not in [target, IDcol]]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(data[predictors1])
data[predictors1] = sc.transform(data[predictors1])
data
```

```python
data_lr_train = data[data['label']>-1]
data_lr_train
```

```python
data_lr_test = data[data['label']==-1]
data_lr_test
```

```python
data_lr_train.to_csv('data_lr_train.csv',index=False)
data_lr_test.to_csv('data_lr_test.csv',index=False)
```
