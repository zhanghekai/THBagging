```python
train = pd.read_csv('ktrain.csv',encoding='gbk')
test = pd.read_csv('ktest.csv',encoding='gbk')
```

```python
train_id = pd.read_csv('df_id_train.csv',header=None)
train_id.columns=['个人编码','label']
train_id
```

```python
train = pd.merge(train,train_id,on='个人编码',how='left')
```

```python
train
```

```python
only1 = train[train.label==1]
```

```python
only1
```

```python
#违法人的行为数量,即1类人的点数
a =only1.个人编码.value_counts()
nums = pd.DataFrame(a)
nums.columns=[['counts']]
nums['个人编码'] = nums.index
nums.index = range(a.shape[0])
nums
```

```python
test_id = pd.read_csv('df_id_test.csv',header=None)
test_id.columns=['个人编码']
test_id['label'] = -1
allId = pd.concat([train_id,test_id],ignore_index=True)
```

```python
test['label']=-1
```

```python
trainAndTest = pd.concat([train,test],ignore_index=True)
trainAndTest
```

```python
import numpy as np
import time
```

```python
predictors = [x for x in train.columns if x not in ['顺序号','个人编码','label']]
```

```python
start = time.time()
for Id in allId.个人编码[:1]:
    #提取的特征，均是与label为1的比较
    maxXiangsiduPer = 0.0#每个点，相似点1中，占某个人的最大比例
    maxXiangsidu = 0#每个点，相似点1中，占某个人的最大个数
    sumtips = 0#每个点，与其相似的1的点的个数的总值
    perSumTips = 0#平均每个点
    sumPerson = 0#每个点，与其相似的1 的人数的总值
    perSumPerson = 0#平均每个点
    maxtip = -1#每个点，与其相似的1的点的个数的最大值
    maxPerson = -1#每个点，与其相似的1的人数的最大值
    targetPerson = trainAndTest[trainAndTest.个人编码==Id]
    tipsNum = targetPerson.shape[0]#每个人的点数
    otherPerson = only1[only1.个人编码!=Id]
    for i in targetPerson.index:
        cha = (otherPerson[predictors]-targetPerson[predictors].loc[i])**2
        cha['sum'] = np.sqrt(cha.T.sum())
        cha['个人编码'] = otherPerson['个人编码']
        cha['顺序号'] = otherPerson['顺序号']
        cha['label'] = otherPerson['label']
        Index = cha['sum'].sort_values(ascending=True)[:1000].index
        top1000 = cha[cha.index.isin(Index)]#最相似的1000个点
        num1 = top1000[top1000.label==1].shape[0]#1的点的个数
        sumtips = sumtips+num1#1的点的总数
        numPerson1 = top1000[top1000.label==1].个人编码.value_counts().shape[0]#1的人的个数
        sumPerson = sumPerson + numPerson1#1的人的总数
        if maxtip<num1:#更新最大值
            maxtip = num1
        if maxPerson<numPerson1:
            maxPerson = numPerson1
            
        persons = pd.DataFrame(top1000[top1000.label==1].个人编码.value_counts())#计算相似点1中，占某个人的最大比例，与最大个数
        persons.columns=['sum']
        persons['个人编码'] = persons.index
        persons.index=range(top1000[top1000.label==1].个人编码.value_counts().shape[0])
        baifenbi = pd.merge(persons,nums,on='个人编码',how='left')
        baifenbi['百分比'] = baifenbi['sum']/baifenbi['counts']
        xiangsiduPer = baifenbi['百分比'].max()
        xiangsidu = baifenbi['sum'].max()
        if maxXiangsiduPer<xiangsiduPer:
            maxXiangsiduPer=xiangsiduPer
        if maxXiangsidu<xiangsidu:
            maxXiangsidu=xiangsidu

    perSumTips = sumtips/tipsNum
    perSumPerson = sumPerson/tipsNum
#     temp = result[result['个人编码']==Id]#更新result中特征的值
#     temp['maxXiangsiduPer'] = maxXiangsiduPer
#     temp['maxXiangsidu'] = maxXiangsidu
#     temp['sumtips'] = sumtips
#     temp['perSumTips'] = perSumTips
#     temp['sumPerson'] = sumPerson
#     temp['perSumPerson'] = perSumPerson
#     temp['maxtip'] = maxtip
#     temp['maxPerson'] = maxPerson
#     result[result['个人编码']==Id] = temp
end = time.time()
print('运行时间：')
end-start
```
