```python
trainAndTest = pd.read_csv("trainAndTest.csv",encoding='gbk')
trainAndTest.head()
```

```python
train_id = pd.read_csv('df_id_train.csv',header=None)
test_id = pd.read_csv('df_id_test.csv',header=None)
```

```python
train_id_0 = train_id[train_id[1]==0]
train_id_1 = train_id[train_id[1]==1]
```

```python
hospital = pd.get_dummies(trainAndTest.医院编码,prefix='医院编码')
hospital = pd.concat([trainAndTest.个人编码,hospital],axis=1)
hospital
```

```python
a = pd.DataFrame(trainAndTest.groupby('个人编码')['医院编码'].value_counts())
a.columns=['每个医院就诊次数']
hospitalTimes = a.reset_index()
hospitalTimes
```

```python
hospitalList = trainAndTest.groupby('个人编码')['医院编码'].unique().reset_index()
```

```python
hospitalList['医院数量'] = hospitalList['医院编码'].apply(lambda x:x.tolist().__len__())
```

```python
hospitalList['顺序号数量'] = trainAndTest.groupby('个人编码')['医院编码'].count().reset_index()['医院编码']
```

```python
hospitalList['平均医院数量'] = hospitalList['医院数量']/hospitalList['顺序号数量']
```

```python
hospitalList['单个医院就诊最多次数'] = hospitalTimes.groupby('个人编码')['每个医院就诊次数'].max().reset_index()['每个医院就诊次数']
hospitalList['单个医院就诊最少次数'] = hospitalTimes.groupby('个人编码')['每个医院就诊次数'].min().reset_index()['每个医院就诊次数']
hospitalList['单个医院就诊平均次数'] = hospitalTimes.groupby('个人编码')['每个医院就诊次数'].mean().reset_index()['每个医院就诊次数']
```

```python
hospitalList['单个医院就诊最多次数/顺序号数量'] = hospitalList['单个医院就诊最多次数']/hospitalList['顺序号数量']
hospitalList['单个医院就诊最少次数/顺序号数量'] = hospitalList['单个医院就诊最少次数']/hospitalList['顺序号数量']
```

```python
detail = pd.read_csv('fee_detail.csv')
detail
```

```python
detailSan = pd.get_dummies(detail.三目统计项目,prefix='三目统计项目')
detailSan['顺序号'] = detail['顺序号']
```

```python
predictors3 = [x for x in detailSan.columns if x not in ['顺序号']]
```

```python
san = pd.DataFrame(detailSan.groupby('顺序号')[predictors3].sum()).reset_index()
```

```python
san
```

```python
danjia_max = pd.DataFrame(detail.groupby('顺序号').单价.max()).reset_index()
danjia_max.columns=['顺序号','单价_max']

danjia_min = pd.DataFrame(detail.groupby('顺序号').单价.min()).reset_index()
danjia_min.columns=['顺序号','单价_min']

danjia_mean = pd.DataFrame(detail.groupby('顺序号').单价.mean()).reset_index()
danjia_mean.columns=['顺序号','单价_mean']

danjia_var = pd.DataFrame(detail.groupby('顺序号').单价.var()).reset_index()
danjia_var.columns=['顺序号','单价_var']

shuliang_max = pd.DataFrame(detail.groupby('顺序号').数量.max()).reset_index()
shuliang_max.columns=['顺序号','数量(三目)_max']

shuliang_min = pd.DataFrame(detail.groupby('顺序号').数量.min()).reset_index()
shuliang_min.columns=['顺序号','数量(三目)_min']

shuliang_mean = pd.DataFrame(detail.groupby('顺序号').数量.mean()).reset_index()
shuliang_mean.columns=['顺序号','数量(三目)_mean']

shuliang_var = pd.DataFrame(detail.groupby('顺序号').数量.var()).reset_index()
shuliang_var.columns=['顺序号','数量(三目)_var']

danjia_sum = pd.DataFrame(detail.groupby('顺序号').单价.sum()).reset_index()
danjia_sum.columns=['顺序号','单价_sum']

shuliang_sum = pd.DataFrame(detail.groupby('顺序号').数量.sum()).reset_index()
shuliang_sum.columns=['顺序号','数量(三目)_sum']

count = pd.DataFrame(detail.groupby('顺序号').size()).reset_index()
count.columns=['顺序号','每个顺序号数量三目项目个数']
```

```python
danjia_var = danjia_var.fillna(0)
shuliang_var = shuliang_var.fillna(0)
```

```python
sanMean = detailSan.groupby('顺序号')[predictors3].mean().reset_index()
sanMean.rename(columns=lambda x:x+'_mean', inplace=True)
sanMean.rename(columns={'顺序号_mean': '顺序号'}, inplace=True)
sanMean
```

```python
result = pd.merge(san,sanMean,on='顺序号')
result = pd.merge(result,danjia_max,on='顺序号')
result = pd.merge(result,danjia_min,on='顺序号')
result = pd.merge(result,danjia_mean,on='顺序号')
result = pd.merge(result,danjia_var,on='顺序号')
result = pd.merge(result,danjia_sum,on='顺序号')
result = pd.merge(result,shuliang_max,on='顺序号')
result = pd.merge(result,shuliang_min,on='顺序号')
result = pd.merge(result,shuliang_mean,on='顺序号')
result = pd.merge(result,shuliang_var,on='顺序号')
result = pd.merge(result,shuliang_sum,on='顺序号')
result = pd.merge(result,count,on='顺序号')
result
```
